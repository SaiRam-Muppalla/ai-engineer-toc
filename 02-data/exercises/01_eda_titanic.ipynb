# EDA on the Titanic Dataset

Perform a complete exploratory data analysis and extract insights.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset (available via seaborn)
df = sns.load_dataset('titanic')
print('Shape:', df.shape)
df.head()
```

```python
# Step 1: Basic info
print(df.info())
print('\nMissing values:\n', df.isnull().sum())
print('\nStatistics:\n', df.describe())
```

```python
# Step 2: Survival rate analysis
print('Overall survival rate:', df['survived'].mean().round(3))

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Survival by sex
df.groupby('sex')['survived'].mean().plot(kind='bar', ax=axes[0], color=['coral', 'steelblue'])
axes[0].set_title('Survival Rate by Sex')
axes[0].set_ylabel('Survival Rate')

# Survival by class
df.groupby('pclass')['survived'].mean().plot(kind='bar', ax=axes[1], color='teal')
axes[1].set_title('Survival Rate by Class')

# Age distribution
df['age'].dropna().plot(kind='hist', bins=30, ax=axes[2], color='purple', alpha=0.7)
axes[2].set_title('Age Distribution')

plt.tight_layout()
plt.show()
```

```python
# Step 3: Correlation heatmap
plt.figure(figsize=(10, 6))
numeric_cols = df.select_dtypes(include=[np.number]).columns
sns.heatmap(df[numeric_cols].corr(), annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()

# TODO: What features are most correlated with survival?
# Write your findings as a comment below:
# ...
```
