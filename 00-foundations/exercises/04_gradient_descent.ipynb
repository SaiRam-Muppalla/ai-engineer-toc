# Exercise 4: Gradient Descent from Scratch

Implement gradient descent to minimize a simple function.

```python
import numpy as np
import matplotlib.pyplot as plt

# We want to minimize f(x) = x^2
# The derivative is f'(x) = 2x

def f(x):
    return x ** 2

def gradient(x):
    return 2 * x

# Gradient descent
x = 10.0          # starting point
learning_rate = 0.1
n_iterations = 50
history = [x]

for _ in range(n_iterations):
    # TODO: Update x using gradient descent rule: x = x - lr * gradient(x)
    x = x - learning_rate * gradient(x)
    history.append(x)

print(f'Final x: {x:.6f} (should be close to 0)')
print(f'Final f(x): {f(x):.6f}')

# Plot convergence
plt.plot(range(len(history)), [f(xi) for xi in history])
plt.xlabel('Iteration')
plt.ylabel('f(x) = x²')
plt.title('Gradient Descent Convergence')
plt.grid(True)
plt.show()
```

```python
# Challenge: Try gradient descent on a 2D function
# Minimize f(x, y) = x^2 + y^2

def f2d(params):
    x, y = params
    return x**2 + y**2

def gradient_2d(params):
    x, y = params
    return np.array([2*x, 2*y])

params = np.array([5.0, 3.0])
lr = 0.1
losses = []

for _ in range(100):
    losses.append(f2d(params))
    params = params - lr * gradient_2d(params)

print(f'Final params: {params}')
print(f'Final loss: {f2d(params):.8f}')
```
